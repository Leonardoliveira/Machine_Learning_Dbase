```
'''Treinando um modelo de classificação para sabe o que é cachorro e o que é porco.
                                    Features(1- sim, 0- não)
Caracteristicas

Pelo longo?
Perna curta?
Faz "au-au"?
'''

#criando porcos
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

#criando cachorros
cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

#juntando os anamais
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]

#definindo as classes (1- porco, 0- cachorro)
treino_y = [1,1,1,0,0,0]
```

```
#Importando bibliotecas
from sklearn.svm import LinearSVC

#instanciando a classe
model = LinearSVC()

#ensinando 
model.fit(treino_x, treino_y)
```

```
#fazendo teste
animal_misterioso = [1, 1, 1,]
model.predict([animal_misterioso])
```

```
#fazendo mais testes
animal1 = [1, 1, 1]
animal2 = [1, 1, 0]
animal3 = [0, 1, 1]

#fazendo a predição
model.predict([animal1, animal2, animal3])

#outra forma
teste_x = [animal1, animal2, animal3]
#fazendo mais teste de classes
teste_y = [0, 1, 1]
previsoes = model.predict(teste_x)
previsoes

```

#Resultado

animal1 = cachorro, 

animal2 = porco, 

animal3 = porco, 

Resultado do modelo

[0, 1, 0] = cachorro, porco, cachorro

ou seja, o modelo teve uma potuação de 2/3.


```
#fazendo comparações para verificar a taxa de acerto
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_de_acerto = corretos / total
print("A taxa de acerto do modelo é: %.2f" %(taxa_de_acerto * 100), end='%.\n')
```

```
#importando o metodo de acuracia
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(teste_y, previsoes)
print("A taxa de acerto do modelo é: %.2f" %(taxa_de_acerto * 100), end='%.\n')
```
